```python
#for translating english to telugu(Meher) using SINGLE SHOT PROMPTING
from transformers import pipeline

# Initialize the translation pipeline with a meher model it translate english text to telugu
translator = pipeline("translation_en_to_te", model="Meher2006/english-to-telugu-model")

# Translate a sample text to telugu
english_text = "This is a test sentence to translate from English to Telugu."
telugu_translation = translator(english_text)

# Output the result in telugu
print(telugu_translation[0]['translation_text'])
```

```python
#for translating english to hindi(Helsinki) using FEW SHOT PROMPTING
from transformers import pipeline

# Load the translation pipeline with English to Hindi with a helsinki model it transalte english to hindi
translator = pipeline("translation_en_to_hi", model="Helsinki-NLP/opus-mt-en-hi")

# Input English text
english_text = """Translate english to hindi
English:How are you?
Hindi:कैसे हो?
English:What is your name?
Hindi:आपका नाम क्या है?
English:who are you?
Hindi:Answer in hindi
"""

# Translate to Hindi
hindi_translation = translator(english_text)

# Print the translated text
print(hindi_translation[0]['translation_text'])
```
